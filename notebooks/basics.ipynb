- In general, hedge funds don't like high beta. High beta means your return is highly dependent upon the market. Alpha means market neurral return

**Curve Building**

- Curve building refers to constructing a smooth curve from market data points. Two majour approaches are:
    1. Exact curve building (exact fit, bootstrapping)
    2. Spline-based curve building (smooth approximation)
- IB and discretionary macro funds need exact curve buildings. Traders care a lot about the spcific dates on the curve.
- You need to understand why you decide to use certain instruments to build a curve - i.e., MUFG uses futures till 1 year point and after 1 year uses SOFR swaps to build a SOFR curve, why?
- Intraday curve tends to be build with a variety of instruments, specific dates as traders use them for hedging, whereas end of day curve has less instruments as stability is more important.
- What is the difference between same SOFR curve but using different instruments? Difference between swap curve vs bond curve?
- For every point in the curve, why you decide to use that specific instrument?
